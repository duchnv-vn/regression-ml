```python
import matplotlib.pyplot as plt
import numpy as np

from load_image import load_images, prepend_bias, load_label, encode_target_label

DIGIT = 1

X_train = load_images("train-images-idx3-ubyte.gz")
Y_train = load_label("train-labels-idx1-ubyte.gz")

print(X_train.shape)
print(Y_train.shape)
```

```python
# Extract examples of target DIGIT
digits = X_train[Y_train.reshape(-1) == DIGIT]
np.random.shuffle(digits)
print(digits.shape)
```

```python
rows, columns = 3, 15
fig = plt.figure(figsize=(10, 4))
for i in range(rows * columns):
    ax = fig.add_subplot(rows, columns, i + 1)
    ax.axis('off')
    ax.imshow(digits[i].reshape((28, 28)), cmap="grey")
plt.show()
```

```python
from binary_clarification import train, test

w = train(prepend_bias(X_train), encode_target_label(Y_train, DIGIT), iter=100, lr=1e-5)
```

```python
print(w)
```

```python
X_test = load_images("t10k-images-idx3-ubyte.gz")
Y_test = load_label("t10k-labels-idx1-ubyte.gz")

print(X_test.shape)
print(Y_test.shape)
```
