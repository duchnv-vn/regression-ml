```python
import csv
import numpy as np

data_list = list(csv.reader(open("Sonar.csv")))

data_without_bias = np.array(data_list)
data_without_bias = data_without_bias[1:, ]  # Remove header
np.random.seed(1234)
np.random.shuffle(data_without_bias)
print(data_without_bias.shape)
```

```python
X = data_without_bias[:, :-1].astype(np.float32)
print(X.shape)
```

```python
Y = data_without_bias[:, -1:].astype(np.int8)
print(Y.shape)
```

```python
sample_size = X.shape[0]
train_size = round(sample_size * 0.8)

X_train, X_test = np.vsplit(X, [train_size])
Y_train_unencoded, Y_test = np.vsplit(Y, [train_size])
print(X_train.shape)
print(Y_train_unencoded.shape)
```

```python
print(X_test.shape)
print(Y_test.shape)
```
